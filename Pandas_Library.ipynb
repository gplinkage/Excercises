```python
def func_sum(*args):
  result = 0
  for i in args:
    result += i
  return result
```

```python
func_sum(2,3)
```

```python
func_sum(2,3,4)
```

```python
def func_sum(**kwargs):
  return kwargs
```

```python
func_sum(2,3)
```

```python
def func_add(*args):
  result=0
  for i in args:
    result= result+i
  return result
```
